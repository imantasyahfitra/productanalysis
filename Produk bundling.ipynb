```python
import pandas as pd
df = pd.read_csv("datasiap.csv")
```

```python
df['Grouped'] = df.groupby('InvoiceNo')['Description'].transform(lambda x: ','.join(x))
```

```python
df2 = df[['InvoiceNo','Grouped']].drop_duplicates()
df2.head(100)
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

list3 = [[2, 3], [2, 3, 4]]

count = Counter()

for sublist in list3:
    count.update(Counter(combinations(sublist, 2)))

print(count)
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

```

```python
count = Counter()
# ini row split biar kita dapet list,kan kalo ga di split tetep satu list ga kayak yang di stackoverflow

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
# angka dua buat mau berpaa pirnya
count.most_common(10)
# for key,value in count.most_common(10):
#     print(key, value)
```
